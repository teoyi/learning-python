```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt 
```

## Creating Data Sets

```python
from sklearn.datasets import make_regression
n = 1000 # Number of data points 
x, y = make_regression(
    n_samples = n,
    n_features = 1,
    n_informative = 1,
    n_targets = 1,
    noise = 100,
    random_state = 1
)
```

## Implementing my own least squares fit function

```python
sigma = np.ones(n)
x = x.reshape(1000)
# print (x1)
# print(y)

S = np.sum(1/(sigma**2))
Sx = np.sum(x/sigma**2)
Sxx = np.sum((x**2)/sigma**2)
Sy = np.sum(y/sigma**2)
Sxy = np.sum((x*y)/sigma**2)
#print( S, Sx, Sxx, Sy, Sxy)
delta = (S * Sxx) - (Sx**2) 
#print(S)
a = (Sy*Sxx - Sx*Sxy)/delta
b = (S*Sxy - Sx*Sy)/delta 
# print(b)
# print(a)
# print(delta)
plt.plot(x, a + b*x, 'r')
plt.scatter(x,y)


```

## Using numpy lib for linear regression

```python
x1 = x.reshape(len(x))
A = np.vstack([x1, np.ones(len(x))]).T
m, c, rank, s = np.linalg.lstsq(A, y, rcond = None)

plt.plot(x, y, 'o', label = 'dataset markers')
plt.plot(x, m + c*x, 'r')
plt.show()
```

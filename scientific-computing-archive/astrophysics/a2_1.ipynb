```python
import numpy as np 
import matplotlib.pyplot as plt 
import os
```

```python
v_0 = 1
#v = np.arange(0,2) 
A = 1
B = 0.01 
x = np.arange(0.0001, 20, 0.001)
s_v = A * x**2 * np.exp(-x)
i_0 = A/x
i_t = A/x + A*x**2 * np.exp(-x)*(1-np.exp(B/x))

plt.loglog(x, s_v, linestyle='--', label = 'Source function of cloud', color = 'r')
plt.loglog(x, i_0, linestyle='--', label = 'Specific Intensity of Star', color = 'orange')
plt.loglog(x, i_t)
plt.ylim([10**-9,10**5])
plt.axvline(x = 0.01, linestyle='--', color = 'black', label = r'$\nu_\tau$')
plt.xlabel(r'$v/v_0$')
plt.ylabel('Specific Intensity (A = 1)')
plt.legend()
plt.savefig('itried.png')
plt.show()

print(i_0)
```

```python
x1 = np.arange(0, 300)
x2 = np.arange(0, 100)
x3 = np.arange(0, 100)
x4 = np.arange(0, 300)
x5 = np.arange(0, 300)
x6 = np.arange(0, 300)

y1 = x1 
y2 = np.exp(-x2/10)
y3 = 2*np.exp(-x3/5)
y4 = x4**4 * np.exp(-x4/10)
y5 = 1/x5 * np.exp(-x5/10)
y6 = 1/(x6+x6**2)

plt.figure() 
plt.loglog(x1,y1, label =r'y = x')
plt.title('a)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_a.png')

plt.figure() 
plt.loglog(x2,y2, label = r'$y = e^{-\frac{x}{10}}$')
plt.loglog(x3, y3, label = r'$y = 2e^{-\frac{x}{5}}$')
plt.title('b)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_b.png')


plt.figure() 
plt.loglog(x4,y4, label = r'$y = x^4e^{-\frac{x}{10}}$')
plt.title('c)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_c.png')


plt.figure()
plt.loglog(x5,y5, label = r'$y = \frac{1}{x}e^{-\frac{x}{10}}$')
plt.title('d)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_d.png')


plt.figure()
plt.loglog(x6,y6, label = r'$y = \frac{1}{x+x^2}$')
plt.title('e)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('a22_e.png')

plt.show()
```
